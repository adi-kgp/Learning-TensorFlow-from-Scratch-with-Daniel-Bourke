What we have created so far:

* Scalar: a single number
* Vector: a number with direction (e.g. wind speed and direction)
* Matrix: a 2-dimensional array of numbers
* Tensor: an n-dimensional array of numbers (where n can be any number , a 0-dimensional tensor is a scalar, a 1-dimensional tensor is a vector)

### Creating random tensors

Random tensors are tensors of some arbritrary size which contain random numbers. 

```python
# Create two random (but the same) tensors
random_1 = tf.random.Generator.from_seed(7) # set seed for reproducibility
random_1 = random_1.normal(shape=(3, 2))
random_1
```

```python
random_2 = tf.random.Generator.from_seed(7)
random_2 = random_2.normal(shape=(3,2))
# Are they equal?
random_1, random_2, random_1 ==random_2
```

### Shuffle the order of elements in a tensor

```python
# Shuffle a tensor (valuable for when you want to shuffle your data so the inherent order does not affect learning)
not_shuffled = tf.constant([[10,7],
                           [3,4],
                           [2,5]])
```

```python
# Shuffle our non-shuffled tensor
tf.random.set_seed(42) # global level random seed
tf.random.shuffle(not_shuffled, seed=42)  # operation level random seed
```

> Rule 4: "If both the global and the operation seed are set: Both seeds are used in conjunction to determine the random sequence."

**Matrix Multiplication**

The rule for matrix multiplication was first formulated by French polymath Jacques Phillip Marie Binet 

<img src="https://upload.wikimedia.org/wikipedia/commons/a/af/Jacques_Binet.jpg">

In machine learning, matrix multiplication is one of the most common tensor operations. We can visualize matrix multiplication on this website: http://matrixmultiplication.xyz/

There are two rules our tensors (or matrices) need to fulfill if we are going to matrix multiply them:

1. The inner dimensions must match. 
2. The resulting matrix has the shape of the outer dimensions.

```python
# Matrix Multiplication in tensorflow
print(tensor)
tf.matmul(tensor, tensor)
```

```python
tensor * tensor  # element wise matrix multiplcation
```

```python
# Matrix multiplication with python operator "@", similar to tf.matmul
tensor @ tensor
```

```python
# Create a tensor of shape (3, 2)
X = tf.constant([[1,2],
                 [3,4],
                 [5,6]])

# Create another (3,2) tensor 
Y = tf.constant([[7, 8],
                 [9, 10],
                 [11, 12]])
X, Y
```

```python
# Try to matrix multiply tensors of same shape
# tf.matmul(X,Y) # InvalidArgumentError: Matrix size-incompatible: In[0]: [3,2], In[1]: [3,2] [Op:MatMul]
```

```python
# Let's change the shape of Y
tf.reshape(Y, shape=(2, 3)), Y
```

```python
X.shape, tf.reshape(Y, shape=(2,3)).shape
```

```python
# Try to matrix multiply X by reshaped Y
X @ tf.reshape(Y, shape=(2,3))
```

```python
tf.matmul(X, tf.reshape(Y, shape=(2,3)))
```

```python
# Try change the shape of X instead of Y
tf.matmul(tf.reshape(X, shape=(2,3)), Y)
```

```python
tf.reshape(X, shape=(2,3)).shape, Y.shape
```

```python
# Can do the same with transpose
tf.transpose(X), tf.reshape(X, shape=(2,3)), X
```

```python
# Try matrix multiplication with transpose rather than reshape
tf.matmul(tf.transpose(X), Y)
```

**The dot product**

Matrix multiplication is also referred to as the dot product.

You can perform matrix multiplication using:
* `tf.matmul()`
* `tf.tensordot()`
* `@`

```python
X, Y
```

```python
# Perform the dot product on X and Y (requires X or Y to be transposed)
tf.tensordot(tf.transpose(X), Y, axes=1)
```

```python
# Perform matrix multiplication between X and Y (transposed)
tf.matmul(X, tf.transpose(Y))
```

```python
# Performs matrix multiplication between X and Y (reshaped)
tf.matmul(X, tf.reshape(Y, shape=(2,3)))
```

```python
# Check the values of Y , reshape Y and transposed Y
print("Normal Y:")
print(Y, "\n") # "\n" is for newline
print("Y reshaped to (2,3):")
print(tf.reshape(Y, (2,3)), "\n")
print("Y transposed:")
print(tf.transpose(Y))
```

Generally when performing matrix multiplication on two tensors and one of the axes does not line up, you will tranpose (rather than reshape one of the tensors to satisfy the matrix multiplication rules).

### Changing the datatype of a tensor

```python
# Create a new tensor with default datatype (float32)
B = tf.constant([1.7, 7.4])
B, B.dtype
```

```python
C = tf.constant([7, 10])
C, C.dtype
```

```python
# Change from float32 to float16 (reduced precision)
D = tf.cast(B, dtype=tf.float16)
D, D.dtype
```

```python
# Change from int32 to float32
E = tf.cast(C, dtype=tf.float32)
E, E.dtype 
```

```python
E_float16 = tf.cast(E, dtype=tf.float16)
E_float16, E
```

### Aggregating tensors

Aggregating tensors  = condensing them from multiple values down to a smaller amount of values.

```python
# Getting the absolute values
D = tf.constant([-7, -10])
D, tf.abs(D)
```

Let's go through the following forms of aggregation:

* Get the minimum
* Get the maximum
* Get the mean of a tensor
* Get the sum of a tensor

```python
# Creating a random tensor with values between 0 and 100 of size 50
E = tf.constant(np.random.randint(0, 100, size=50))
E
```

```python
tf.size(E), E.shape, E.ndim
```

```python
# Find the minimum
tf.reduce_min(E), np.min(E)
```

```python
# Find the maximum
tf.reduce_max(E), np.max(E)
```

```python
# Find the mean
tf.reduce_mean(E)
```

```python
# Find the sum 
tf.reduce_sum(E)
```

```python
# Find the variance of our tensor , we need access to tensorflow_probability
import tensorflow_probability as tfp
tfp.stats.variance(E), tf.math.reduce_variance(tf.cast(E, dtype=tf.float32))
```

```python
# Find the standard deviation
tfp.stats.stddev(tf.cast(E, dtype=tf.float32)), tf.math.reduce_std(tf.cast(E, dtype=tf.float32))
```

**Find the positional maximum and minimum**

```python
# Create a new tensor for finding positional minimum and maximum
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
# Find the positional maximum
tf.argmax(F), tf.argmin(F)
```

```python
# Index on our largest value position
F[tf.argmax(F)]
```

```python
# Find the max value of F
tf.reduce_max(F)
```

```python
# Check for equality
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
# Find the positional minimum
tf.argmin(F)
```

```python
# Find the minimum using the positional minimum index
F[tf.argmin(F)], tf.reduce_min(F)
```
